```python
import pandas as pd
import numpy as np
```

<ul><strong>
    Đọc dữ liệu trong diemthi2017
    </strong>
</ul>

```python
diemthi2017 =  pd.read_csv('./dataset/diemthi2017.csv')
diemthi2017
```

<ul>
    <strong>
    Tạo dict chứa các tỉnh theo mã tỉnh
    </strong>
</ul>

```python
temp=diemthi2017[['SOBAODANH','Tỉnh Thành']].copy()# lấy dữ liệu hai cột SOBAODANH và tỉnh thành
temp['SOBAODANH']=np.floor((temp['SOBAODANH']/10**6)) # lấy phần nguyên
new_dict = dict(zip(temp['SOBAODANH'].unique(), temp['Tỉnh Thành'].unique())) #tạo dict chưa  mã tỉnh thành và tỉnh thành
new_dict
```

<ul><strong>Tạo hàm thêm tỉnh thành vào dataset</strong></ul>

```python
def add_tinhthanh(dataset,feature):
    location= []
    location = np.floor(dataset[feature]/10**6)
    locaion=location.astype(int)
    for i in range(len(location)):
        location[i]=new_dict[location[i]]
    dataset['Tỉnh Thành']=location
```

<ul>
    <strong>
    Tạo hàm lưu xuất dữ liệu vào file csv
    </strong>
</ul>

```python
def xuat_file(dataframe,filename):
    dataframe.to_csv(filename, sep=',',encoding='utf-8-sig')
```

<ul>
    <strong>
        Thêm cột 'Tỉnh thành' vào hai file dataset 2020 và 2022
    </strong>
</ul>

```python
diem2020_df =pd.read_csv('./dataset/diemthi2020.csv')
add_tinhthanh(diem2020_df,'sbd')
xuat_file(diem2020_df,"./dataset/diemthi2020.csv")
```

```python
diem2022_df =pd.read_csv('./dataset/diemthi2022.csv')
add_tinhthanh(diem2022_df,'sbd')
xuat_file(diem2022_df,"./dataset/diemthi2022.csv")
```
